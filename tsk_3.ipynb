TRAINING AND TESTING

```python
datashow = pd.concat([train_data,test_data])
```

```python
le = LabelEncoder()
for column in datashow.columns:
    if datashow[column].dtype == 'object':
        datashow[column] = le.fit_transform(datashow[column])
```

```python
X_train, X_test = train_test_split(datashow,test_size = 0.2, random_state = 42)
```

```python
X_train.dtypes
```

```python
if 'datetime_column' in X_train.columns:
    X_train['datetime_column'] = pd.to_datetime(X_train['datetime_column']).astype(int) // 10**9
```

```python
X_train = pd.get_dummies(X_train)
print(X_train.dtypes)
```

```python
from sklearn.ensemble import IsolationForest

# Initialize the Isolation Forest model
isf = IsolationForest()

# Fit the model with the transformed data
isf.fit(X_train)
```

```python
X_test['y_pred'] = isf.predict(X_test)
```

```python
X_test['y_pred'] = X_test['y_pred'].replace({-1: 1, 1: 0})
X_test.head()
```

```python
y_pred = X_test['y_pred']
is_fraud = X_test['is_fraud']

report = classification_report(is_fraud, y_pred)

print(report)
```
